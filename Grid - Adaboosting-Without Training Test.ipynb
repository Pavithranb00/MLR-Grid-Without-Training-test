```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age','bmi','children','sex_male','smoker_yes']]
```

```python
independent
```

```python
dependent=dataset[['charges']]
```

```python
dependent
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import AdaBoostRegressor
params_grid={'random_state':[0,1,10,1000],
             'n_estimators':[1,10,100,1000,2000],
             'learning_rate':[1.0],
             'loss':['linear','square','square']}
grid=GridSearchCV(AdaBoostRegressor(),params_grid,refit=True,verbose=3,n_jobs=-1)
                    
grid.fit(independent,dependent)
```

```python
# print best parameter after tuning
#print(grid.best_params_)
re=grid.cv_results_



print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```

```python
table
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=int(input("Sex Male 0 or 1:"))
smoker_yes_input=int(input("Smoker Yes 0 or 1:"))
```

```python
Future_Prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])

#change the parameter ,play with it
print("Future_Prediction={}".format(Future_Prediction))
```
